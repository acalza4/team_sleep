```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('REM_NREM.csv')
```

```python
df.shape
```

```python
cols_to_norm = list(df.columns[1:])

ct = ColumnTransformer(transformers = [('min_max_scaler', MinMaxScaler(), cols_to_norm)], remainder = 'passthrough')
df_norm = pd.DataFrame(ct.fit_transform(df), columns = df.columns)
```
